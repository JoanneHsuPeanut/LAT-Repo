### 簡化資料
單獨取出款項名稱與金額，將金額依照款項名稱分組，加總金額，加上index

```python
pd.read_csv(file)[['Complex2Name', 'FValue']].groupby(['Complex2Name'])['FValue'].sum().reset_index()
```

### 取得資料開始與截止年

```python
Yfrom = int(input('資料的開始年'))
Yend = int(input('資料最後一年'))

year = []
for i in range(Yfrom,Yend+1):
    year.append(str(i))
```

### 串接每一年的結果

```python
### 第一筆
result = pd.read_csv(year[0]+'.csv')[['Complex2Name', 'FValue']].groupby(['Complex2Name'])['FValue'].sum().reset_index()
result['year'] = year[0]

### concat 剩下年份
for j in range(1,len(year)):
    new = pd.read_csv(year[j]+'.csv')[['Complex2Name', 'FValue']].groupby(['Complex2Name'])['FValue'].sum().reset_index()
    new['year'] = year[j]
    result = pd.concat([result,new])

result
```
