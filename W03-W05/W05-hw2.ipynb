## dataframe

```python
import pandas as pd
```

### 預覽資料

```python
file = ['./data/PISASCIENCE.csv','./data/PISAMATH.csv','./data/PISAREAD.csv']
df = pd.read_csv(file[0])

df.head(3)
```

### 簡化資料

```python
df = pd.read_csv(file[0])[["TIME","LOCATION","SUBJECT","INDICATOR","Value"]]
df
```

### 串接三個科目

```python
#concat all

for i in range(1,len(file)):
    new = pd.read_csv(file[i])[["TIME","LOCATION","SUBJECT","INDICATOR","Value"]]
    df = pd.concat([df,new])

df
```

```python
import plotly_express as px
```

## 沒啥意義的圖

```python
fig = px.scatter(df, x="TIME", y="Value", color="INDICATOR", marginal_y="violin",
           marginal_x="box", trendline="ols", template="simple_white")
fig.show()
```

```python
fig = px.box(df, x="INDICATOR", y="Value", color="TIME", notched=True)
fig.show()
```

## 可能有點用的圖

```python
fig = px.box(df, x="TIME", y="Value", color="INDICATOR", notched=True)
fig.show()
```

```python
fig = px.violin(df, y="Value", x="INDICATOR", color="SUBJECT", box=True, points="all", hover_data=df.columns)
fig.show()
```

```python
mask = df["TIME"] == 2018
dfyear = df[mask]
dfyear
```

## 以下為不嚴謹的三個問題圖解

```python
year = [2006,2009,2012,2015,2018]


for i in year:
    mask = (df["TIME"] == i) & (df["SUBJECT"] != "TOT")
    dfyear = df[mask]
    fig = px.violin(dfyear, y="Value", x="INDICATOR", color="SUBJECT", box=True, points="all", hover_data=df.columns, title=i)
    fig.show()
```

### Ｑ1 迷思：「文組就女生的強項」
圖中可見每一年的閱讀成績女生都比男性高出一大截，可能是真的

### Q2 迷思：「女生數學爛很合理」
圖中可見每一年的數學成績女生並沒有明顯低於男生，所以還是乖乖算數學吧

### Q3 迷思：「理組是男生的強項」
圖中可見每一年的科學成績男女幾乎表現相同，所以想選理組就給他選下去
